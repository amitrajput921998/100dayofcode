# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,(200))
```

```python
y=np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my Title")
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y)
```

```python
color=np.random.rand(50)
sizes=1000*np.random.random(50)
plt.scatter(x,y,c=color,s=sizes,alpha=1)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
color=np.random.rand(50)
sizes=1000*np.random.random(50)
plt.scatter(x,y,c='r',s=sizes,alpha=1)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```
